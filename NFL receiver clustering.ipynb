```python
#Here, we will remove the values that aren't really associated with being a receiver. These will include defense measures
#and throwing measures.
#We are removing these variables so we can better focus on the differences in wide reveiver.
wr1 = wr[['Name', 'position', 'ovr rating', 'awareness rating',
       'kick return rating', 'lead block rating',
       'strength rating', 'b c vision rating', 'catch in traffic rating',
       'pursuit rating', 'medium route running rating',
       'catching rating', 'acceleration rating', 'spin move rating', 'height',
       'finesse moves rating', 'spectacular catch rating', 'run block rating',
       'injury rating', 'weight',
       'deep route running rating', 'trucking rating',
       'throw accuracy short rating', 'juke move rating',
       'play recognition rating', 'short route running rating',
       'speed rating', 'run block power rating',
       'jumping rating', 'toughness rating',
       'stiff arm rating', 'power moves rating',
       'release rating', 'hit power rating',
       'pass block power rating',
       'impact blocking rating', 'stamina rating', 'carrying rating',
       'break tackle rating', 'press rating',
       'block shedding rating', 'run block finesse rating', 'agility rating', 'elusiveness rating']]
```

```python
#To test our receiver theory, we will try the experiment with just 2 clusters
cluster = KMeans(n_clusters=2)
wr1['cluster'] = cluster.fit_predict(wr1.iloc[:, 2:])
wr1['cluster'].value_counts()
```

```python
cluster02 = wr1[wr1['cluster'] == 0]
cluster12 = wr1[wr1['cluster'] == 1]
```

```python
cluster02_stats = cluster02.describe()
cluster12_stats = cluster12.describe()
```
